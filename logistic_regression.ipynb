```python
# Step 1: Imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sn
import matplotlib.pyplot as plt

# Step 2: Read CSV (no header=None since file has header)
data = pd.read_csv('diabetes.csv')

# Step 3: Features and label
feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']
x = data[feature_cols]
y = data['label']

# Step 4: Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)

# Step 5: Train model
model = LogisticRegression(solver='lbfgs', max_iter=1000)
model.fit(x_train, y_train)

# Step 6: Predict
y_pred = model.predict(x_test)

# Step 7: Evaluate
conf_mat = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', conf_mat)

acc = accuracy_score(y_test, y_pred)
print('Accuracy Score:', acc)
print('Accuracy in Percentage:', int(acc * 100), '%')

# Step 8: Plot heatmap
conf_df = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(conf_df, annot=True, fmt='d')
plt.show()
```
